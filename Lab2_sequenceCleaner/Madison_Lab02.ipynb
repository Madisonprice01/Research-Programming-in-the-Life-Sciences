### seqCleaner

```python
#!/usr/bin/env python3 
# Name: Madison Price (maeprice@ucsc.edu)
# Group Members: Kiana Imani (kimani) and Queenie Li (qli86)

'''
Read a DNA string from user input and return a collapsed substring of embedded Ns to: {count}.

Example: 
 input: AaNNNNNNGTC
output: AA{6}GTC

Any lower case letters are converted to uppercase

Assumptions: The only letters the user might enter are A, C, T, G (upper or lowercase), and N (uppercase)
Design Decisions: I decided to delete the unnecessary class method, length() 
'''

class DNAstring (str):
    '''Build a DNAstring object from a string.'''
    def purify(self):
        ''' Return an upcased version of the string, collapsing a single run of Ns.'''
        purified = []                                     # initiate a list for the unambigious letters
        count = 0                                         # initiate a count variable
        for i in self:                    
            if i != 'N':                                  # two possibilities for nonambiguous letters
                if count == 0:                            # 1st possibility = no ambiguous letters have been encountered yet
                    purified.append(i)                    # append unambiguous letter to the list
                if count != 0:                            # 2nd possibility = we have reached an unambiguous letter after a block of ambiguous letters
                    purified.append('{'+str(count)+'}')   # append the count of ambiguous letters to the list
                    purified.append(i)                    # append unambiguous letter to the list
                    count = 0                             # reset count variable so it isn't added to the list more than once
            if i == 'N':                                  # if letter is ambigious add it to the count variable
                count += 1
        if count >= 1 and len(purified) != 0:             # test for case of N's at end of sequence
            purified.append('{'+str(count)+'}')
        if count >= 1 and len(purified) == 0:             # test for case of only N's
             purified.append('{'+str(count)+'}')
        purifiedString = ''.join(purified)                # turn DNA sequence as a list into a string
        upcasedPurifiedString = purifiedString.upper()    # make sure string is all uppercase
        return upcasedPurifiedString                      
    
def main():
    ''' Get user DNA data and clean it up using DNAstring class.'''
    data = input('Enter a DNA sequence: ')
    thisDNA = DNAstring (data)
    pureData = thisDNA.purify()
    print (pureData)
    
main()
```

```python
#!/usr/bin/env python3 
# Name: Madison Price (maeprice)
# Group Members: Kiana Imani (kimani) and Queenie Li (qli86)

'''
“Parse” sequence name information from a single line of a FASTQ formatted file, where the line is obtained from user input.

input: seqname line of a FASTQ file
output: parsed out fields of the run information

Assumptions: The input string has seven fields (separated by colons) and starts with "@"
Design Decisions: I used concatenation within print() instead of format methods
'''

class FastqString (str):
    ''' Create a FastqString object from a string.'''
    def parse(self):
        ''' Parse self into fields that are displayed on separate lines.'''
        fields = self[1:]                              # eliminate "@"
        splitFields = fields.split(":")                # split input line around colons and add fields to a list
        print("Instrument = "+splitFields[0])          # print each field on a separate line with its own introduction
        print("Run ID = "+splitFields[1])
        print("Flow Cell ID = "+splitFields[2])
        print("Flow Cell Lane = "+splitFields[3])
        print("Tile Number = "+splitFields[4])
        print("X-coord = "+splitFields[5])
        print("Y-coord = "+splitFields[6])
        
def main():
    '''Get user seqname line data and format it using FastqString class.'''
    seqnameLineData = input("Enter seqname line: ")
    fastqStringObject = FastqString (seqnameLineData)
    fastqStringObject.parse()
    
main()
```

(Note: make sure that the angle returned is in **degrees** !!)
### Hints: 
 - Each coordinate will contain only 3 numeric values. 
 - Bond lengths are the distance between two points, which for points P and Q in 3-space,
$(P_x , P_y , P_z )$ and $(Q_x , Q_y , Q_z )$ respectively, the distance between them is:


$$
\begin {eqnarray}
     \|PQ\| &=&\sqrt{(P_x-Q_x)^2 + (P_y-Q_y)^2 + (P_z-Q_z)^2)} \\
            &=&\sqrt{\sum^{i \in x,y,z}{(P_i-Q_i)^2}} \\
\end{eqnarray}
$$

 - Bond angles can be calculated from the dot product.
 
Let’s say that we have three points in space labeled P, Q and R. We are interested in the angle at point Q that is made from line segments QP and QR.
The dot product tells us (for standard vectors P and R) that:
$$P \cdot R=\|P\| \|R\| \cos{\theta}$$
in this notation,$\|P\|$ refers to the length of vector P as a standard vector (assumed to begin at the origin (0,0,0) ).
We can then see that the angle between vectors P and R can be found by:
$$\cos{\theta}=\frac{P \cdot R} {\|P\| \|R\|}$$

We can calculate the dot product using the sum of products of the vector components:
$$P \cdot R = \sum^{i \in x,y,z}{P_i R_i} $$

Now, to find vector P in standard form, we need to remember that QP starts at Q, so we need to place the origin at Q and find out where P is in that new space. We do that by subtracting the components of Q from P.
Putting all of this together, we get:
$$\theta=\cos^{-1}{\frac{\sum^{i \in x,y,z}{(P_i - Q_i) (R_i - Q_i) }} {\|QP\| \|QR\|}}$$
Remember, $\theta$ is in radians.

```python
#!/usr/bin/env python3
# Name: Madison Price (maeprice)
# Group Members: Kiana Imani (kimani) and Queenie Li (qli86)

'''
Take three sets of atomic coordinates and calculate bond lengths and bond angles.

input: three sets of atomic coordinates
output: two bond lengths and one bond angle

Assumptions: The input line contains three sets of coordinates, each of which contains 3 numeric values.
Design Decisions: I deal with the single line of input by splitting it around any parentheses to yield individual strings with atomic coordinates.
                  I used eval() to convert the strings to tuples to be used by the Triad class.
                  I convert from radians to degrees using a single unit conversion.
'''

import math


class Triad:
    """
    Calculate angles and distances among a triad of points.

    Author: David Bernick
    Date: March 21, 2013
    Points can be supplied in any dimensional space as long as they are consistent.
    Points are supplied as tupels in n-dimensions, and there should be three
    of those to make the triad. Each point is positionally named as p,q,r
    and the corresponding angles are then angleP, angleQ and angleR.
    Distances are given by dPQ(), dPR() and dQR()

    Required Modules: math
    initialized: 3 positional tuples representing Points in n-space
             p1 = Triad( p=(1,0,0), q=(0,0,0), r=(0,1,0) )
    attributes: p,q,r the 3 tuples representing points in N-space
    methods:  angleP(), angleR(), angleQ() angles measured in radians
          dPQ(), dPR(), dQR() distances in the same units of p,q,r

    """

    def __init__(self, p, q, r):
        """ Construct a Triad.

        Example construction:
            p1 = Triad( p=(1.,0.,0.), q=(0.,0.,0.), r=(0.,0.,0.) ).
        """
        self.p = p
        self.q = q
        self.r = r

    # private helper methods
    def d2(self, a, b):  # calculate squared distance of point a to b
        return float(sum((ia - ib) * (ia - ib) for ia, ib in zip(a, b)))

    def dot(self, a, b):  # dotProd of standard vectors a,b
        return float(sum(ia * ib for ia, ib in zip(a, b)))

    def ndot(self, a, b, c):  # dotProd of vec. a,c standardized to b
        return float(sum((ia - ib) * (ic - ib) for ia, ib, ic in zip(a, b, c)))

    # calculate lengths(distances) of segments PQ, PR and QR
    def dPQ(self):
        """ Provides the distance between point p and point q """
        return math.sqrt(self.d2(self.p, self.q))

    def dPR(self):
        """ Provides the distance between point p and point r """
        return math.sqrt(self.d2(self.p, self.r))

    def dQR(self):
        """ Provides the distance between point q and point r """
        return math.sqrt(self.d2(self.q, self.r))

    def angleP(self):
        """ Provides the angle made at point p by segments pq and pr (radians). """
        return math.acos(
            self.ndot(self.q, self.p, self.r) / math.sqrt(self.d2(self.q, self.p) * self.d2(self.r, self.p)))

    def angleQ(self):
        """ Provides the angle made at point q by segments qp and qr (radians). """
        return math.acos(
            self.ndot(self.p, self.q, self.r) / math.sqrt(self.d2(self.p, self.q) * self.d2(self.r, self.q)))

    def angleR(self):
        """ Provides the angle made at point r by segments rp and rq (radians). """
        return math.acos(
            self.ndot(self.p, self.r, self.q) / math.sqrt(self.d2(self.p, self.r) * self.d2(self.q, self.r)))



def main():
    ''' Get atomic coordinates from user and use Triad class to calculate bond lengths and angles. '''
    coords = input("Enter three sets of atomic coordinates: ")
    atom1 = "C"
    atom2 = "N"
    atom3 = "Ca"

    modifiedCoords = coords.replace(")", "(")          # make sure string contains only one type of paranthesis
    splitCoords = modifiedCoords.split("(")            # split string around that type of paranthesis, and add components to a list
    atom1Coords = eval(splitCoords[1])                 # turn elements of the list into tuples and assign approrpriate names
    atom2Coords = eval(splitCoords[3])
    atom3Coords = eval(splitCoords[5])
    triad = Triad(atom1Coords, atom2Coords, atom3Coords)   # create a Traid object with the tuples (coords)

    print(atom2 + "-" + atom1 + " bond length = " + "{:.2f}".format(triad.dPQ()))                      # print bond lengths and angle using Triad methods
    print(atom2 + "-" + atom3 + " bond length = " + "{:.2f}".format(triad.dQR()))
    print(atom1 + "-" + atom2 + "-" + atom3 + " bond angle = " + "{:.1f}".format(57.3*triad.angleQ())) # convert radians to degrees using a concersion factor

main()
```
